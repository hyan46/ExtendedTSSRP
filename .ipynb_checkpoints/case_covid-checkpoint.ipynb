# Data Processing

```python
data_normal = np.array(data.iloc[:,3:]).T[:80,:]
data_normal = np.diff(data_normal,n=1,axis=0)
data_normal_mean = np.mean(data_normal)
data_normal_std = np.std(data_normal)
data_normal = data_normal - data_normal_mean
data_normal = data_normal/data_normal_std

data_abnormal = np.array(data.iloc[:,3:]).T[80:,:]
data_abnormal = np.diff(data_abnormal,n=1,axis=0)
data_abnormal = data_abnormal - data_normal_mean
data_abnormal = data_abnormal/data_normal_std

data_combined = np.concatenate((data_normal, data_abnormal), axis=0)
```

```python
for i in range(len(data)):
#     plt.plot(np.array(data.iloc[i,10:]) - np.array(data.iloc[i,3:-7]))
    if i in [10, 38]:
        plt.plot(data_combined[:,i], label = data.iloc[i,2])
    else:
        plt.plot(data_combined[:,i], c='grey')
plt.legend()
```

# Create Model Instance

```python
M = np.zeros((data.shape[0],data.shape[0]))
for i in range(len(data)):
    M[i,i] = np.mean(data_abnormal[:,i])

p = data.shape[0] # Number of dimensions
c = 1 # Target meanshift is c * M
Ks = 5   #Select Top Ks control chart
T0 = 50
k = data.shape[0] #Number of failuer Mode
nmodels = 3
M_tssrp = np.kron(np.eye(p),np.ones((1,1))) # Mean Failure Modes
nsensors = Ks # Number of selected sensors
experiment_log = {'c':c, 'Ks':Ks, 'statistics':[[] for i in range(nmodels)],
                  'sens_history':[[] for i in range(nmodels)],
                  'mode_history':[[] for i in range(nmodels)],
                  'seq_stats':[[] for i in range(nmodels)]}

model_all = [[] for i in range(nmodels)]
model_all[0] = ExtendedTSSRP(p, c, k,M,nsensors,Ks,L=-1,chart = 'srp',mode='T2')
model_all[1] = TSSRP(p, c, p,M_tssrp,nsensors,Ks,L=-1,chart='srp',mode='T2')
mumin = 1
model_all[2]= AdaptCUSUM(p, c, p, M, nsensors, nsensors, mumin = mumin,delta=0.1)

# failure_idx = 0
# model_idx = 0
# all_Simulated_data[failure_idx].shape

sequential_statistis_topRsum, sensor_selection_history, failure_mode_history, i,sequential_statistics = model_all[0].compute_monitoring_statistics(data_combined,T0,-1)
```
