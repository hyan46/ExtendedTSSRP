# Learn PPCA

```python
import scipy.io as io
import matplotlib.pyplot as plt
import numpy as np
from ppca import PPCA
```

### Load the data from the Markdown

```python
data = io.loadmat('tonnage.mat')
X = data['X']
y = data['y'][:,0]
np.moveaxis(X,0,1).shape
vx = np.moveaxis(X,0,1).reshape(-1,582).T
y[y==5]=2
plt.plot(vx[0,:])
```

```python
vx[y==0]
```

```python
plt.plot(y)
```

```python
from ppca import PPCA
ppca = PPCA(svd_solver='auto', n_components=10)
ppca.fit(vx)
```

```python
np.mean(ppca.score_samples(vx))
```

### Created Selected Sampling in index matrix

```python
index = [50]
ppca.score_samples_index(vx[[1],:],index)
```
